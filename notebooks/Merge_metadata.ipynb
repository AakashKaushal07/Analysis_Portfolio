```python
import pandas as pd
import os
```

```python
PATH = "D:\All Season Data\Files"
```

```python
all_dfs= {}
for l1,l2,files in os.walk(PATH):
    for file in files:
        # print(f"{l1}\\{file}")
        # print(f"{l2}\\{file}")
        full_path = f"{l1}\\{file}"

        full_path = full_path.replace("\\All Season Data\\Files","")
        full_path = full_path.replace("b1_","")
        full_path = full_path.replace("b2_","")
        full_path = full_path.replace("b3_","")
        full_path = full_path.replace("b4_","")
        full_path = full_path.replace("b5_","")
        full_path = full_path.replace("b6_","")
        full_path = full_path.replace("b7_","")
        full_path = full_path.replace("b8_","")
        full_path = full_path.replace("b9_","")
        
        if file not in all_dfs.keys() :
            all_dfs[full_path] = []
        df = pd.read_excel(f"{l1}\\{file}")
        # print(df.columns)
        if 'event_url' in df.columns :
            df['event_url'] = df['event_url'].astype(str)
            pattern = r'^(http|https|ftp)://[^\s]+$'
            valid_links_df = df.query("event_url.str.match(@pattern)")
        else:
            df['shot_url'] = df['shot_url'].astype(str)
            pattern = r'^(http|https|ftp)://[^\s]+$'
            valid_links_df = df.query("shot_url.str.match(@pattern)")
        all_dfs[full_path].append(valid_links_df)
    
```

```python
true_counts = pd.read_excel("Leagues Counts.xlsx")
```

```python
true_counts.head()
```

```python
league_count_map = {}
for i,row in true_counts.iterrows():
    key = str(row.get('xl_file_name')).split("/")[-1]
    cnt = row.get('File_matches',-1)
    if key not in league_count_map.keys() :
        league_count_map[key] = 0
    league_count_map[key] = cnt
```

```python
league_count_map
```

```python
for x in all_dfs.keys():
    elements = x.split("\\")[:-1]
    for i in range(2,len(elements)+1):
        pt = "\\".join(elements[:i])
        # print())
        os.makedirs(pt,exist_ok=True)
```

```python
merged_items = {}
for k,v in all_dfs.items() :
    if k not in merged_items.keys() :
        merged_items[k] = None
    merged_df = pd.concat(v, ignore_index=True)
    dropped_df = merged_df.drop_duplicates(subset=["home_team","away_team","datetime"])
    dropped_df.to_excel(k,index=False)
    
    if dropped_df.shape[0] > league_count_map.get(k,-1) and dropped_df.shape[0] - league_count_map.get(k,-1) :
        print(f"{k} : Total Len = {merged_df.shape[0]} | Unique len = {dropped_df.shape[0]} | Found Len = {league_count_map.get(k,'NA')}")
```
