```python
match_info = pd.read_csv('match_info.csv')
match_info
```

```python
match_event_data = pd.read_csv('event_data.csv')
match_event_data.head()
```

```python
type_id_df = pd.read_csv('opta-events.csv', encoding = 'ISO-8859-1')
type_id_df.rename(columns={'ID':'typeId'},inplace=True)
type_id_df.head()
```

```python
qualifier_df = pd.read_csv('opta-qualifiers.csv')
qualifier_df.head()
```

```python
id1 = match_info['id'].iloc[0]
team1 = match_info['officialName'].iloc[0]
id2 = match_info['id'].iloc[1]
team2 = match_info['officialName'].iloc[1]

match_event_data['contestantId'] = np.where(match_event_data['contestantId']==id1,team1,team2)
match_event_data['outcome'] = np.where(match_event_data['outcome']==1,'Successful','Unsuccessful')
match_event_data.rename(columns={"contestantId":'Team','timeMin':'Minute','timeSec':'Seconds'},inplace=True)
```

```python
type_cols = [col for col in match_event_data.columns if '/qualifierId' in col]
match_event_data['endX'] = 0.0
match_event_data['endY'] = 0.0

for i in range(len(match_event_data)):
    df1 = match_event_data.iloc[i:i+1,:]
    j = 0
    for j in range(len(type_cols)):
        col = df1[type_cols[j]].values[0]
        if col == 140:
            endx = df1.loc[:,'qualifier/%i/value' %j].values[0]
            match_event_data['endX'][i] = float(endx)
        else:
            j +=1
    k = 0
    for k in range(len(type_cols)):
        col = df1[type_cols[k]].values[0]
        if col == 141:
            endy = df1.loc[:,'qualifier/%i/value' %k].values[0]
            match_event_data['endY'][i] = float(endy)
        else:
            k +=1
```

```python
match_event_data = pd.merge(match_event_data,type_id_df,on='typeId',how='left')
match_event_data.head(8)
```

```python
match_event_data['Goal_Mouth_Y'] = 0.0
match_event_data['Goal_Mouth_Z'] = 0.0

for i in range(len(match_event_data)):
    df1 = match_event_data.iloc[i:i+1,:]
    j = 0
    for j in range(len(type_cols)):
        col = df1[type_cols[j]].values[0]
        if col == 102:
            mouth_y = df1.loc[:,'qualifier/%i/value' %j].values[0]
            match_event_data['Goal_Mouth_Y'][i] = float(mouth_y)
        else:
            j +=1
    k = 0
    for k in range(len(type_cols)):
        col = df1[type_cols[k]].values[0]
        if col == 103:
            mouth_z = df1.loc[:,'qualifier/%i/value' %k].values[0]
            match_event_data['Goal_Mouth_Z'][i] = float(mouth_z)
        else:
            k +=1
    
    #nj=0 Add furthur shot details
    
```

```python
match_event_data['Big Chance'] = 0.0
match_event_data['Own Goal'] = 0.0

for i in range(len(match_event_data)):
    df1 = match_event_data.iloc[i:i+1,:]
    j = 0
    for j in range(len(type_cols)):
        col = df1[type_cols[j]].values[0]
        if col == 214:
            big_chance = 'Yes'
            match_event_data['Big Chance'][i] = big_chance
        else:
            j +=1
    k = 0
    for k in range(len(type_cols)):
        col = df1[type_cols[k]].values[0]
        if col == 28:
            own_g = 'Yes'
            match_event_data['Own Goal'][i] = own_g
        else:
            k +=1
    
    #nj=0 Add furthur shot details
    
```

```python
final_df = match_event_data[['Team','Minute','Seconds','typeId','Type','outcome','x','y','endX','endY','Goal_Mouth_Y'
                             ,'Goal_Mouth_Z','Big Chance','Own Goal','playerName','keyPass','Description']]

final_df['keyPass'].fillna(0,inplace=True)
final_df.dropna(inplace=True)
final_df = final_df.reset_index(drop=True)

os.chdir(output_data_dir)
final_df.to_csv(f"{team1}_vs_{team2}.csv")
```
